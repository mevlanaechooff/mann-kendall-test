```python
import numpy as np
import pandas as pd
import pymannkendall as mk
import matplotlib.pyplot as plt


#zaman serisi oluşturuyoruz
np.random.seed(42)  # Aynı sonuçları elde etmek için sabitlenmiş bir rastgelelik
zaman = pd.date_range('2000-01-01', periods=100, freq='ME') 

veriler = np.random.normal(loc=0, scale=1, size=100).cumsum() + np.arange(100)

# Verileri bir DataFramee koyduk
df = pd.DataFrame({'Zaman': zaman, 'Değer': veriler})
df.set_index('Zaman', inplace=True)

# Veriyi görselleştirdik
df.plot(title="Zaman Serisi Verisi", figsize=(10, 6))
plt.show()


# Test
sonuc = mk.original_test(df['Değer'])

#Sonuçları yazdıralım
print(sonuc)

```

```python
# Yıllık sıcaklık verileri
np.random.seed(42)
yillar = pd.date_range(start='1970', periods=50, freq='Y')
sicaklik = 0.02 * np.arange(50) + np.random.normal(0, 0.5, 50)

# Veriyi bir DataFramee koyalım
df_sicaklik = pd.DataFrame({'Yillar': yillar, 'Sicaklik': sicaklik})
df_sicaklik.set_index('Yillar', inplace=True)

# Veriyi görselleştirelim
df_sicaklik.plot(title="Yıllık Ortalama Sıcaklık", figsize=(10, 6))
plt.show()

# Mann-Kendall testi
sicaklik_sonuc = mk.original_test(df_sicaklik['Sicaklik'])

#Sonuçları yazdıralım
print(sicaklik_sonuc)
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pymannkendall as mk
import yfinance as yf
from scipy.stats import linregress

# Apple hisse senedi verilerini 2010-2020 arasında alıyoruz
df = yf.download('AAPL', start='2010-01-01', end='2024-01-01')
print(df.head())

# Kapanış fiyatlarına odaklanalım
df_close = df['Close']

plt.figure(figsize=(10,6))
sns.lineplot(x=df_close.index, y=df_close.values, marker='o', label="AAPL Kapanış Fiyatı")
plt.title("Apple Hisse Senedi Kapanış Fiyatları (2010-2024)", fontsize=14)
plt.xlabel("Yıl", fontsize=12)
plt.ylabel("Kapanış Fiyatı (USD)", fontsize=12)
plt.grid(True)
plt.legend()
plt.show()

# Mann-Kendall testi uygulama
test_result = mk.original_test(df_close)

# Test sonuçlarını yazdırma
print("Mann-Kendall Trend Testi Sonuçları:")
print(f"Trend Yönü: {test_result.trend}")
print(f"p-Değeri: {test_result.p}")
print(f"Z Değeri: {test_result.z}")
print(f"Kendall's Tau: {test_result.Tau}")
print(f"Eğim (Slope): {test_result.slope}")

from scipy.stats import linregress

#Lineer regresyon işlemi
slope, intercept, r_value, p_value, std_err = linregress(df_close.index.map(pd.Timestamp.toordinal), df_close.values)

#Trend çizgisi ile görselleştirme
plt.figure(figsize=(10,6))
sns.lineplot(x=df_close.index, y=df_close.values, marker='o', label="AAPL Kapanış Fiyatı")
plt.plot(df_close.index, intercept + slope * df_close.index.map(pd.Timestamp.toordinal), 'r', label=f'Trend Çizgisi (Eğim: {slope:.3f})')
plt.title("Apple Hisse Senedi Kapanış Fiyatları ve Trend Çizgisi (2010-2020)", fontsize=14)
plt.xlabel("Yıl", fontsize=12)
plt.ylabel("Kapanış Fiyatı (USD)", fontsize=12)
plt.grid(True)
plt.legend()
plt.show()

```
